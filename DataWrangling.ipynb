#**Data Normalization**

* Simple Feature Scalling
* Min - max
* Z - Score

```python
#simple feature scalling

mydata['width'].head()
```

```python
mydata[['width', 'length', 'height']].head()
```

```python
#simple feature scalling = width / max width
mydata['width'] = mydata['width']/mydata['width'].max()
mydata[['width', 'length', 'height']].head()
```

```python
mydata['width'].describe()
```

```python
# min-max = (length - min length) / (max length - min length)

mydata['length'] = (mydata['length']-mydata['length'].min())/(mydata['length'].max()-mydata['length'].min())

mydata[['width', 'length', 'height']].head()
```

```python
mydata['length'].describe()
```

```python
# Z-score = height - height mean / height standard deviation

mydata['height'] = (mydata['height']-mydata['height'].mean())/mydata['height'].std()

mydata[['width', 'length', 'height']].describe()
```

if you are using one method, all standardize to one method. if not semua akan ada slightly different value. but recommended to use the same method for all. normally people use **z score**.

#**Data Binning**

```python
mydata['price'].describe()   #to understand the data better
```

```python
#ni nak put dia in category.

price_bins = np.linspace(mydata_new1['price'].min(),mydata_new1['price'].max(),num=4)
print (price_bins)

price_groups=['Low Price', 'Mid Price', 'High Price']  #ni kita just put nama category je

mydata_new1['price_bins'] = pd.cut(mydata_new1['price'],bins=price_bins,labels=price_groups,include_lowest=True)
#guna ni untuk link kan category and numbers tu
#include_lowest tu to avoid ambil null value

mydata_new1[['price', 'price_bins']].head(10)
```

```python
mydata_new1['price_bins'].value_counts() #check the categories. these are called imbalance classes (unequal amount in the categories)
#in ML nanti the machine will ambil low price model lebih banyak so jadi bias.
```

```python
mydata_new1.describe()  #takyah include all pun sebab binning only applicable to numerical data. so focus numerical je
#nak buat prediction, boleh try tengok features with difference gap min and max dia besar
```

```python
norm_bins = np.linspace(mydata_new1['normalized-losses'].min(),mydata_new1['normalized-losses'].max(),num=4)
print (norm_bins)

norm_groups=['Low', 'Mid', 'High']

mydata_new1['normalized_losses_bins'] = pd.cut(mydata_new1['normalized-losses'],bins=norm_bins,labels=norm_groups,include_lowest=True)

mydata_new1[['normalized-losses', 'normalized_losses_bins']].head(10)
```
