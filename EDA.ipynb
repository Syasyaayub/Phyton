```python
link = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'

import pandas as pd

df = pd.read_csv(link)

df.head()
```

```python
#create variables to select only numerical features

df_numeric = df.select_dtypes(include =['number'])
df_numeric.corr()
```

```python
#ni kita pilih mana numerical features yang kita nak tengok correlation

df[['engine-size', 'city-mpg', 'bore', 'stroke', 'horsepower', 'price']].corr()
```

```python
#guna regplot untuk see visually the distribution sekali dengan correlation dia

from matplotlib import pyplot as plt
import seaborn as sns

sns.regplot(x = 'engine-size', y = 'price', data = df)

#positive correlation
```

```python
sns.regplot(x = 'city-mpg', y = 'price', data = df)

#negative correlation
```

```python
sns.regplot(x='stroke', y= 'price', data = df)
```

```python
import scipy.stats as st  #guna unruk calculate Pearson method. coeefficient and P-value

correlation, p_value = st.pearsonr(df['engine-size'], df['price'])

print('The correlation is: ', correlation, 'and the p_value is: ', p_value)

#strong certainty of the p_value
```

```python
correlation, p_value = st.pearsonr(df['city-mpg'], df['price'])

print('The correlation is: ', correlation, 'and the p_value is: ', p_value)
```

```python
df['stroke'].isnull().sum()
```

```python
import numpy as np

df['stroke'].replace(np.nan, df['stroke'].mean(), inplace = True)
```

```python
correlation, p_value = st.pearsonr(df['stroke'], df['price'])

print('The correlation is: ', correlation, 'and the p_value is: ', p_value)
```

#**Group By**

```python
df1 = df[['drive-wheels', 'body-style', 'price']]

df1_group = df1.groupby(['drive-wheels', 'body-style'], as_index = False).mean()
#if true dia cam 4wd merge then breakdown body style.
#if false, dia every row letak 4wd tu. senang tengok

df1_group
```

```python
#rows = index. dalam pivot (rows, columns)

df1_pivot = df1_group.pivot(index = 'drive-wheels', columns = 'body-style')

df1_pivot
```

```python
df1_pivot = df1_group.pivot(index = 'body-style', columns = 'drive-wheels')

#change rows and columns je. either one is fine. nak tengok correlation je

df1_pivot
```

```python
#pivot can use to create heatmap

plt.pcolor(df1_pivot, cmap = 'RdBu')

plt.colorbar()
plt.xlabel('Body Style')
plt.ylabel('Drive Wheels')

plt.show()
```

```python
sns.heatmap(df1_pivot, annot = True, cmap ='RdBu')   # True = visible annotation
```

```python
plt.figure(figsize=(18,12))  #normal device max width is 18, height sukahati
sns.heatmap(df_numeric.corr(), annot = True, cmap='coolwarm')  #ni dia ambil semua yang numeric untuk tengok correlation
```

```python
df2 = df[['num-of-doors', 'aspiration', 'price']]

df2_group = df2.groupby(['num-of-doors', 'aspiration'], as_index = False).mean()

df2_pivot = df2_group.pivot(index = 'num-of-doors', columns = 'aspiration')

df2_pivot
```

```python
sns.heatmap(df2_pivot, annot = True, cmap = 'GnBu')
```
